```python
import pandas as pd
```

```python
# Load each sheet into a separate DataFrame
df1 = pd.read_excel('gendered_attack.xlsx', sheet_name='Sheet1')
df2 = pd.read_excel('gendered_attack.xlsx', sheet_name='Sheet2')
df3 = pd.read_excel('gendered_attack.xlsx', sheet_name='Sheet3')
df4 = pd.read_excel('gendered_attack.xlsx', sheet_name='Sheet4')
df5 = pd.read_excel('gendered_attack.xlsx', sheet_name='Sheet5')
```

```python
# Extract column names from each DataFrame
columns_df1 = df1.columns.tolist()
columns_df2 = df2.columns.tolist()
columns_df3 = df3.columns.tolist()
columns_df4 = df4.columns.tolist()
columns_df5 = df5.columns.tolist()
```

```python
# Find the maximum number of columns in any DataFrame to standardize the length
max_length = max(len(columns_df1), len(columns_df2), len(columns_df3), len(columns_df4), len(columns_df5))
```

```python
# Standardize the length of the lists
columns_df1.extend([None] * (max_length - len(columns_df1)))
columns_df2.extend([None] * (max_length - len(columns_df2)))
columns_df3.extend([None] * (max_length - len(columns_df3)))
columns_df4.extend([None] * (max_length - len(columns_df4)))
columns_df5.extend([None] * (max_length - len(columns_df5)))
```

```python
# Create a new DataFrame with these lists as columns
df_columns = pd.DataFrame({
    'df1': columns_df1,
    'df2': columns_df2,
    'df3': columns_df3,
    'df4': columns_df4,
    'df5': columns_df5
})

df_columns
```

#### We made sure that all DFs above share the same column titles

```python
# Add a new column to each DataFrame
df1['new_column'] = 'df1'
df2['new_column'] = 'df2'
df3['new_column'] = 'df3'
df4['new_column'] = 'df4'
df5['new_column'] = 'df5'

# Now each DataFrame has a new column named 'new_column' 
# with all its values set to the respective DataFrame's name
df2
```

```python
import pandas as pd

# Concatenate the DataFrames
giant_df = pd.concat([df1, df2, df3, df4, df5], axis=0)

# Reset the index of the new DataFrame, if necessary
giant_df = giant_df.reset_index(drop=True)

# Now giant_df is a combination of df1, df2, df3, df4, and df5

giant_df
```

```python
# Check for any columns that have all NaN values
columns_with_all_nan = giant_df.columns[giant_df.isna().all()]

# Print the result
if len(columns_with_all_nan) > 0:
    print("Columns with all NaN values:", columns_with_all_nan.tolist())
else:
    print("No columns with all NaN values.")
```

```python
# Identify columns that have all NaN values
columns_to_drop = giant_df.columns[giant_df.isna().all()]

# Drop these columns from giant_df
giant_df.drop(columns_to_drop, axis=1, inplace=True)
```

```python
# Lower-casing all column names
giant_df.columns = giant_df.columns.str.lower().str.replace(' ', '_')
giant_df.columns
```
